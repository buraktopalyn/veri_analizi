```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import seaborn as sns
import pandas as pd
penguins = sns.load_dataset("penguins")
df = penguins.copy()
```

```python
df.head()
```

```python
df["species"].value_counts().plot.barh();
```

## Sütun Grafik

```python
sns.barplot(x="species", y=df.species.index, data=df);
```

```python
sns.catplot(x="species", y="body_mass_g", data=df);
```

```python
sns.barplot(x="species", y="body_mass_g", hue="island", data=df);
```

## Histogram ve Yoğunluk Grafikleri

```python
sns.distplot(df.body_mass_g, kde=False, bins=50);
```

```python
sns.distplot(df.body_mass_g, kde=True, bins=30);
```

```python
sns.kdeplot(df.body_mass_g, shade= True);
```

```python
(sns
.FacetGrid(df, hue="species", height=5)
 .map(sns.kdeplot, "body_mass_g", shade=True)
 .add_legend()
);
```

```python
sns.catplot(x="species", y="body_mass_g", hue="island", kind="strip", data=df);
```

## Kutu Grafik

```python
sns.boxplot(x=df.body_mass_g);
```

```python
sns.boxplot(x="island", y="body_mass_g", hue="sex", data=df);
```

## Violin Grafik

```python
sns.catplot(y="body_mass_g", kind="violin", data=df, color="Yellow");
```

```python
sns.catplot(x="island", y="bill_depth_mm", hue="sex", kind="violin", data=df);
```

## Korelasyon Grafikleri

```python
sns.scatterplot(x="body_mass_g", y="flipper_length_mm", data=df);
```

```python
sns.scatterplot(
    x="body_mass_g", 
    y="bill_length_mm", 
    hue="species", 
    style="island", 
    size="body_mass_g", 
    data=df);
```

```python
sns.lmplot(
    x="body_mass_g", 
    y="flipper_length_mm", 
    hue="island", 
    col="species", 
    data=df);
```

```python
sns.pairplot(df);
```

```python
sns.pairplot(df, hue="species", markers=["o","s","D"]);
```

```python
sns.pairplot(df, kind="reg", hue="species");
```

## Isı Haritası

```python
sns.heatmap(
    df.corr(numeric_only=True), 
    cmap="Greens", 
    annot=True, 
    vmin=-1, 
    vmax=1);
```
