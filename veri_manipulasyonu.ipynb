```python
df2
```

```python
pd.concat([df,df2], ignore_index=True, join="inner")
```

```python
dff1 = pd.DataFrame({
    "Eleman": list("ABCDEFG"),
    "Departman": list("XYYXZXY"),
    "Maaş": np.random.randint(10000,30000,size=7)
})
dff2 = pd.DataFrame({
    "Eleman": list("ABCDEFG"),
    "Puan": np.random.randint(1,100,size=7),
})
```

```python
dff1
```

```python
dff2
```

```python
dfff = pd.merge(dff1, dff2, on="Eleman")
dfff
```

```python
dff3 = pd.DataFrame({
    "Departman": list("ZYX"),
    "Renk": ["Mavi", "Kırmızı", "Yeşil"],
})
```

```python
dff3
```

```python
df = pd.merge(dfff, dff3)
df
```

# Gruplama İşlemleri

```python
import seaborn as sns
```

```python
df = sns.load_dataset("penguins")
```

```python
df.head()
```

```python
df.groupby("species")["body_mass_g"].mean()
```

```python
df.groupby("island")["body_mass_g"].describe()
```

```python
df.iloc[:,1:6].groupby("island").aggregate(["min", np.mean, max])
```

```python
df.iloc[:,1:6].groupby("island").filter(lambda x: x["body_mass_g"].mean() > 3800)
```

```python
df.iloc[:,2:-2] = df.iloc[:,2:-2].transform(lambda x: x*10)
```

```python
df.head()
```

```python
df.iloc[:,2:-1].apply(np.sqrt)
```

```python
df.iloc[:,2:-1].apply(lambda x: x-x.std())
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
df.groupby(["sex","class"])[["survived"]].aggregate("mean").unstack()
```

```python
df.pivot_table("survived", index="sex", columns=["class"])
```

```python
age = pd.cut(df.age, [0,18,40,100])
```

```python
df.pivot_table("survived", ["sex", age], "class")
```
